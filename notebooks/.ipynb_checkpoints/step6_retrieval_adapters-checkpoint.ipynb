# Step 6 — Retrieval Adapters (Part A: Logic Tests)

**Part A** validates retrieval adapter logic without any HTTP calls or model loading.  
All tests run in the `nikko` conda environment (Python 3.11, pydantic, requests).  

**Part B** (live HTTP) requires network access and is tested separately.

Adapters covered:
- `CachedBaseAdapter` — cache key, TTL, HEAD-check logic
- `BetterHealthAdapter` + `WHOAdapter` — static corpus keyword scoring
- `PubMedAdapter` + `HealthdirectAdapter` — import smoke test + RetrievalError structure

**Spec refs:** SPEC-200 §5.4, REQ-200-071, REQ-200-ER1–ER5, REQ-200-160

```python
# ── Setup ────────────────────────────────────────────────────────────
import sys, os, json, hashlib
from pathlib import Path
from datetime import datetime, timedelta, timezone

# Add repo root to sys.path so imports like 'from docs.schemas...' work.
REPO_ROOT = Path('D:/Git Repos/nikko-companion').resolve()
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

print(f'Repo root: {REPO_ROOT}')
print(f'Python: {sys.version}')
```

```python
# ── Test 1: Import smoke test ────────────────────────────────────────
# Verifies all adapter modules import cleanly and ADAPTER_PRIORITY_ORDER
# has exactly 4 entries in the correct order. (REQ-200-071)

from retrieval import (
    ADAPTER_PRIORITY_ORDER,
    PubMedAdapter,
    HealthdirectAdapter,
    BetterHealthAdapter,
    WHOAdapter,
)

expected_order = [PubMedAdapter, HealthdirectAdapter, BetterHealthAdapter, WHOAdapter]
assert ADAPTER_PRIORITY_ORDER == expected_order, (
    f'Priority order mismatch: {ADAPTER_PRIORITY_ORDER}'
)
print('PASS  Import smoke test — all 4 adapters imported, priority order correct')
```

```python
# ── Test 2: Cache key determinism ───────────────────────────────────
# Same (source, query) must always produce the same SHA-256 hex key.
# Different sources or different queries must produce different keys.

from retrieval.base_adapter import CachedBaseAdapter

# Instantiate a concrete adapter to call _cache_key on.
bh  = BetterHealthAdapter()
who = WHOAdapter()

# Determinism: same query → same key.
k1 = bh._cache_key('anxiety')
k2 = bh._cache_key('anxiety')
assert k1 == k2, 'Cache key is non-deterministic!'

# Normalisation: leading/trailing spaces and case are stripped before hashing.
k3 = bh._cache_key('  Anxiety  ')
assert k1 == k3, f'Case/whitespace normalisation failed: {k1} != {k3}'

# Different source → different key (even same query).
k4 = who._cache_key('anxiety')
assert k1 != k4, 'Different sources produced the same cache key!'

# Key is a valid 64-char hex string (SHA-256 output).
assert len(k1) == 64 and all(c in '0123456789abcdef' for c in k1)

print(f'PASS  Cache key determinism — key sample: {k1[:16]}...')
```

```python
# ── Test 3: TTL validity check ──────────────────────────────────────
# is_cache_valid() must return True for fresh items and False for expired ones.
# Uses the adapter's CACHE_POLICY.ttl_days. (REQ-200-ER5)

from docs.schemas.acp_schemas import EvidenceItem, EvidenceTier, SourceTier

now = datetime.now(timezone.utc)

def make_item(retrieved_at):
    return EvidenceItem(
        title='Test article',
        abstract='Test abstract',
        url='https://example.com/test',
        source_name='Test Source',
        evidence_tier=EvidenceTier.GREY_LITERATURE,
        source_tier=SourceTier.PRIMARY,
        retrieved_at=retrieved_at,
    )

# BetterHealth TTL = 30 days.
# Fresh item (29 days ago) → should be valid.
fresh_item = make_item(now - timedelta(days=29))
assert bh.is_cache_valid(fresh_item), 'Fresh item (29d) should be valid.'

# Stale item (31 days ago) → should be invalid.
stale_item = make_item(now - timedelta(days=31))
assert not bh.is_cache_valid(stale_item), 'Stale item (31d) should be invalid.'

# Boundary: exactly at TTL boundary is treated as expired (< not <=).
boundary_item = make_item(now - timedelta(days=30))
# This may be valid or invalid depending on sub-second timing; skip strict check.
# Just verify no exception is raised.
_ = bh.is_cache_valid(boundary_item)

print('PASS  TTL validity check — fresh=True, stale=False')
```

```python
# ── Test 4: HEAD check interval ─────────────────────────────────────
# _should_head_check() must return False for PubMed (no HEAD check policy)
# and True/False correctly for grey-lit adapters based on retrieved_at. (REQ-200-ER5)

from retrieval.pubmed_adapter import PubMedAdapter
pm = PubMedAdapter()

# PubMed has head_check_interval=None → _should_head_check always False.
recent_pubmed = make_item(now - timedelta(days=1))
old_pubmed    = make_item(now - timedelta(days=10))
assert not pm._should_head_check(recent_pubmed), 'PubMed should never HEAD-check'
assert not pm._should_head_check(old_pubmed),    'PubMed should never HEAD-check'

# BetterHealth HEAD interval = 7 days.
# Item retrieved 6 days ago → HEAD check NOT due.
recent_bh = make_item(now - timedelta(days=6))
assert not bh._should_head_check(recent_bh), '6d-old item should not trigger HEAD check'

# Item retrieved 8 days ago → HEAD check IS due.
old_bh = make_item(now - timedelta(days=8))
assert bh._should_head_check(old_bh), '8d-old item should trigger HEAD check'

print('PASS  HEAD check interval — PubMed never, BetterHealth at 7d')
```

```python
# ── Test 5: Static adapter keyword scoring ───────────────────────────
# Build a synthetic StaticCacheIndex with known entries, inject it into
# BetterHealthAdapter, and verify that search() ranks by keyword score.

import tempfile
from docs.schemas.retrieval_schemas import StaticCacheIndex, StaticCacheEntry, StaticCacheQueryParams

# Synthesise two articles.
entry_a = {
    'title': 'Understanding Anxiety Disorders',
    'url': 'https://www.betterhealth.vic.gov.au/health/conditionsandtreatments/anxiety',
    'source_name': 'Better Health Channel',
    'body_markdown': 'Anxiety disorders are a group of mental health conditions that cause excessive fear.',
    'last_fetched': '2026-03-01',
    'cache_expires_at': '2026-04-01',
    'keywords': ['anxiety', 'fear', 'panic', 'mental health'],
}
entry_b = {
    'title': 'Depression: Signs and Symptoms',
    'url': 'https://www.betterhealth.vic.gov.au/health/conditionsandtreatments/depression',
    'source_name': 'Better Health Channel',
    'body_markdown': 'Depression is a mood disorder. It causes persistent sadness and loss of interest.',
    'last_fetched': '2026-03-01',
    'cache_expires_at': '2026-04-01',
    'keywords': ['depression', 'sadness', 'mood', 'mental health'],
}

synthetic_index = {
    'source_name': 'Better Health Channel',
    'last_updated': '2026-03-01',
    'entries': [entry_a, entry_b],
}

# Write to a temp file and load via adapter.
with tempfile.NamedTemporaryFile(mode='w', suffix='.json', delete=False) as f:
    json.dump(synthetic_index, f)
    tmp_path = Path(f.name)

test_adapter = BetterHealthAdapter(index_path=tmp_path)
assert len(test_adapter._index.entries) == 2, 'Index should have 2 entries'

# Query 'anxiety' → entry_a should rank first (title + keyword match).
result = test_adapter.search(StaticCacheQueryParams(query='anxiety', max_results=5))
assert len(result.items) >= 1, 'Expected at least 1 result for anxiety query'
assert 'Anxiety' in result.items[0].title, (
    f'Expected anxiety article first, got: {result.items[0].title}'
)
assert result.grey_literature_flag is True, 'grey_literature_flag must be True'
assert result.cache_hit is True, 'Static adapter should always report cache_hit=True'

# Query 'depression' → entry_b should rank first.
result2 = test_adapter.search(StaticCacheQueryParams(query='depression mood', max_results=5))
assert len(result2.items) >= 1, 'Expected at least 1 result for depression query'
assert 'Depression' in result2.items[0].title, (
    f'Expected depression article first, got: {result2.items[0].title}'
)

# Query with no match → empty results, not an error.
result3 = test_adapter.search(StaticCacheQueryParams(query='quantum physics', max_results=5))
assert len(result3.items) == 0, 'Out-of-scope query should return 0 results'

# Cleanup temp file.
tmp_path.unlink()

print('PASS  Static adapter keyword scoring — ranking and empty-result case correct')
```

```python
# ── Test 6: Empty index graceful degradation ─────────────────────────
# When the static index file has no entries, search() must return an empty
# RetrievalResult — not a RetrievalError and not an exception. (REQ-200-160)

# Default BetterHealth adapter loaded with the actual stub file (empty entries).
default_bh = BetterHealthAdapter()
r = default_bh.search(StaticCacheQueryParams(query='anxiety', max_results=5))

from docs.schemas.retrieval_schemas import RetrievalResult
assert isinstance(r, RetrievalResult), f'Expected RetrievalResult, got {type(r)}'
assert len(r.items) == 0, 'Empty index should produce 0 items'

print('PASS  Empty index graceful degradation — returns RetrievalResult with 0 items')
```

```python
# ── Test 7: RetrievalError structure ────────────────────────────────
# Verify that RetrievalError is a Pydantic model with the correct fields.
# The Evidence Retrieval Agent relies on this to handle failures gracefully.
# (REQ-200-160)

from docs.schemas.retrieval_schemas import RetrievalError

err = RetrievalError(
    source_name   = 'PubMed',
    error_code    = 'timeout',
    error_message = 'ESearch timed out after 10s',
    retryable     = True,
    occurred_at   = datetime.now(timezone.utc),
)

assert err.source_name == 'PubMed'
assert err.error_code  == 'timeout'
assert err.retryable   is True
assert isinstance(err.occurred_at, datetime)

# Non-retryable error.
err2 = RetrievalError(
    source_name   = 'PubMed',
    error_code    = 'parse_error',
    error_message = 'Malformed XML from EFetch',
    retryable     = False,
    occurred_at   = datetime.now(timezone.utc),
)
assert err2.retryable is False

print('PASS  RetrievalError structure — all fields validated')
```

```python
# ── Test 8: PubMed XML parser ───────────────────────────────────────
# Test _parse_pubmed_xml() against a minimal synthetic PubMed XML snippet.
# This validates the parser in isolation without any HTTP calls.

from retrieval.pubmed_adapter import PubMedAdapter

SYNTHETIC_XML = '''<?xml version="1.0" encoding="UTF-8"?>
<PubmedArticleSet>
  <PubmedArticle>
    <MedlineCitation>
      <PMID>12345678</PMID>
      <Article>
        <ArticleTitle>Cognitive Behavioural Therapy for Anxiety Disorders</ArticleTitle>
        <Abstract>
          <AbstractText>CBT is an evidence-based treatment for anxiety disorders.</AbstractText>
        </Abstract>
        <Journal>
          <JournalIssue>
            <PubDate><Year>2023</Year><Month>05</Month></PubDate>
          </JournalIssue>
        </Journal>
      </Article>
    </MedlineCitation>
  </PubmedArticle>
  <PubmedArticle>
    <MedlineCitation>
      <PMID>87654321</PMID>
      <Article>
        <ArticleTitle>Structured Abstract Example</ArticleTitle>
        <Abstract>
          <AbstractText Label="BACKGROUND">Background text here.</AbstractText>
          <AbstractText Label="CONCLUSIONS">Conclusions here.</AbstractText>
        </Abstract>
        <Journal>
          <JournalIssue>
            <PubDate><Year>2022</Year></PubDate>
          </JournalIssue>
        </Journal>
      </Article>
    </MedlineCitation>
  </PubmedArticle>
</PubmedArticleSet>'''

pm = PubMedAdapter()
retrieved_at = datetime.now(timezone.utc)
items = pm._parse_pubmed_xml(SYNTHETIC_XML, retrieved_at)

assert len(items) == 2, f'Expected 2 items, got {len(items)}'

# Article 1: plain abstract.
a1 = items[0]
assert a1.url == 'https://pubmed.ncbi.nlm.nih.gov/12345678/'
assert 'Cognitive' in a1.title
assert 'CBT' in a1.abstract
assert a1.publication_date == '2023-05'

# Article 2: structured abstract (labelled sections concatenated).
a2 = items[1]
assert a2.url == 'https://pubmed.ncbi.nlm.nih.gov/87654321/'
assert 'BACKGROUND:' in a2.abstract
assert 'CONCLUSIONS:' in a2.abstract
assert a2.publication_date == '2022'

print('PASS  PubMed XML parser — plain and structured abstracts, date extraction')
```

```python
# ── Test 9: Adapter class constants ─────────────────────────────────
# Verify SOURCE_NAME, SOURCE_TIER, EVIDENCE_TIER, CACHE_POLICY are correctly
# set on each adapter class (not instance). These are used by the Evidence
# Retrieval Agent to build EvidencePayload. (SPEC-200 §5.4, REQ-200-071)

from docs.schemas.acp_schemas import EvidenceTier, SourceTier

# PubMed — peer-reviewed.
assert PubMedAdapter.SOURCE_NAME   == 'PubMed Central Open Access'
assert PubMedAdapter.SOURCE_TIER   == SourceTier.PRIMARY
assert PubMedAdapter.EVIDENCE_TIER == EvidenceTier.PEER_REVIEWED
assert PubMedAdapter.CACHE_POLICY.ttl_days == 7
assert PubMedAdapter.CACHE_POLICY.head_check_interval is None

# Healthdirect — grey-lit, weekly HEAD check.
assert HealthdirectAdapter.SOURCE_NAME   == 'Healthdirect Australia'
assert HealthdirectAdapter.EVIDENCE_TIER == EvidenceTier.GREY_LITERATURE
assert HealthdirectAdapter.CACHE_POLICY.ttl_days == 30
assert HealthdirectAdapter.CACHE_POLICY.head_check_interval.days == 7

# BetterHealth — grey-lit.
assert BetterHealthAdapter.SOURCE_NAME   == 'Better Health Channel'
assert BetterHealthAdapter.EVIDENCE_TIER == EvidenceTier.GREY_LITERATURE
assert BetterHealthAdapter.CACHE_POLICY.ttl_days == 30

# WHO — grey-lit.
assert WHOAdapter.SOURCE_NAME   == 'World Health Organization'
assert WHOAdapter.EVIDENCE_TIER == EvidenceTier.GREY_LITERATURE
assert WHOAdapter.CACHE_POLICY.ttl_days == 30

print('PASS  Adapter class constants — SOURCE_NAME, tier, cache policy all correct')
```
